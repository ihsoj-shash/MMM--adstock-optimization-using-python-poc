```python
def advertising_after_adstock(adstock):
    dataframe[['Advertising_after_adstock']]=0
    dataframe[['squared_error']]=0
    for cell_loc in range(1, dataframe.shape[0]):
        dataframe.loc[cell_loc,'Advertising_after_adstock'] = dataframe.loc[cell_loc,'Advertising_before_adstock']+(adstock*dataframe.loc[cell_loc-1,'Sales'])
        dataframe.loc[cell_loc,'squared_error'] = (dataframe.loc[cell_loc,'Advertising_before_adstock']-dataframe.loc[cell_loc,'Advertising_after_adstock'])**2
    #sse=(dataframe[['squared_error']]).sum()
    return dataframe[['squared_error']].sum()/dataframe.shape[0]
```

```python
from scipy.optimize import least_squares
res=[]
found=[]
if len(res)!=0:
    res.clear()
if len(found)!=0:
    found.clear()
for ad_val in ad_list:
    result = least_squares(advertising_after_adstock, [ad_val])
    res.append(result.cost)
    found.append(result.x)
```
